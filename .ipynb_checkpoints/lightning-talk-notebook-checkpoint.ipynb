# Topic 1 : Air Quality Prediction Using Environmental Monitoring Data

## Problem Statement

Air pollution is a major environmental and public health concern in urban areas. High concentrations of pollutants such as PM2.5 can negatively impact human health, reduce visibility, and contribute to environmental degradation. Accurate prediction of air quality levels can help government agencies and environmental organizations implement proactive measures to improve air quality and protect public health.

The objective of this project is to develop a machine learning model capable of predicting PM2.5 concentration levels using environmental and meteorological measurements collected from multiple monitoring stations.

### Goal

- Develop a machine learning model to predict PM2.5 concentration levels using environmental and meteorological data.
- Identify the key factors that most strongly influence air quality and pollution levels.
- Evaluate the performance of different machine learning algorithms for air quality prediction.

## Approach

1. Combine the 12 station CSV files into one dataset
2. Handle missing values in CO, NO2, SO2, O3, and PM2.5 using forward-fill
3. Encode categorical features (wind direction, station)
4. EDA: target distribution, correlations, outliers
5. Train and compare regression models (e.g. Linear Regression, Random Forest)
6. Evaluate using RMSE and R²

## Data Info
**Data Source:** [Multi-Site Air-Quality Data Set – Kaggle](https://www.kaggle.com/datasets/sid321axn/beijing-multisite-airquality-data-set?select=PRSA_Data_Dingling_20130301-20170228.csv)

# Topic 3 : Motor Fault Detection using Motor Current Signature Analysis (MCSA)

## Problem Statement

Induction motors are critical components in industrial facilities, and 
unexpected motor failures can cause costly downtime and safety hazards. 
Two of the most common fault types — Broken Rotor Bar (BRB) and Stator 
Winding Fault (SWF) — are difficult to detect visually and typically 
require intrusive inspection methods.

Motor Current Signature Analysis (MCSA) offers a non-invasive approach 
by analysing the three-phase current drawn by the motor to identify 
fault-related frequency patterns. However, traditional rule-based MCSA 
systems rely on fixed thresholds that may not generalise across different 
operating conditions or motor types.

The objective of this project is to build a machine learning classification 
model that can automatically identify the motor condition (healthy, BRB, or 
SWF) from raw three-phase current measurements collected at multiple motor 
speeds — replacing manual threshold tuning with a data-driven approach.

## Goals

- Build a classification model to predict motor condition 
  (healthy / BRB / SWF) from three-phase current signals
- Identify which signal features and operating speeds have the 
  strongest influence on fault classification
- Compare the performance of at least two machine learning models
- Evaluate whether a data-driven approach can generalise across 
  multiple motor speeds (50%, 75%, 100%)

## Approach

1. Combine raw three-phase current CSV files into one master dataset
2. EDA: target distribution, feature types, null values, signal analysis
3. Apply sliding window feature extraction to compute RMS, kurtosis,
   peak, and standard deviation per phase per window
4. Train and compare classification models (e.g. Random Forest, SVM)
5. Evaluate using accuracy, precision, recall, and F1 score per class

  ## Data Info
**Data Source:**[Motor Current and Vibration Monitoring Dataset for various Faults in an E-motor-driven Centrifugal Pump](https://doi.org/10.4121/2b61183e-c14f-4131-829b-cc4822c369d0)

```python
# MISSING VALUES ANALYSIS
# Count missing values
df3.isnull().sum()
```

```python
# Since target is categorical, we compare signal amplitude distributions
# across conditions using RMS (root mean square) per condition group

signal_cols = ['A_m1', 'B_m1', 'C_m1', 'A_m2', 'B_m2', 'C_m2',
               'A_m3', 'B_m3', 'C_m3', 'A_m4', 'B_m4', 'C_m4',
               'A_m5', 'B_m5', 'C_m5']

rms_by_condition = df3.groupby('condition')[signal_cols].apply(
    lambda x: np.sqrt((x**2).mean())
)

print("Mean RMS per Signal Column by Condition:")
display(rms_by_condition.T)

# Plot RMS comparison
fig, axes = plt.subplots(1, 3, figsize=(15, 4), sharey=True)
conditions = ['healthy', 'brb', 'swf']
colors     = ['#2ecc71', '#e74c3c', '#3498db']

for ax, cond, color in zip(axes, conditions, colors):
    vals = rms_by_condition.loc[cond]
    ax.bar(range(len(vals)), vals.values, color=color, edgecolor='black')
    ax.set_title(f'{cond.upper()} — RMS per Signal')
    ax.set_xlabel('Signal Column')
    ax.set_ylabel('RMS Amplitude')
    ax.set_xticks(range(len(vals)))
    ax.set_xticklabels(vals.index, rotation=90, fontsize=7)

plt.suptitle('Mean RMS Signal Amplitude by Condition', fontsize=13, y=1.02)
plt.tight_layout()
plt.show()
```

```python
# ==========================================
# CORRELATION WITH TARGET VARIABLE
# MCSA Dataset
# ==========================================

target = 'condition'

# Encode target numerically for correlation analysis
# healthy = 0, brb = 1, swf = 2
condition_map = {'healthy': 0, 'brb': 1, 'swf': 2}

# Use a sample to avoid memory issues with 2.7M rows
sample_df3 = df3.sample(n=50000, random_state=42).copy()
sample_df3['condition_encoded'] = sample_df3[target].map(condition_map)

# Signal columns only
signal_cols = ['A_m1', 'B_m1', 'C_m1', 'A_m2', 'B_m2', 'C_m2',
               'A_m3', 'B_m3', 'C_m3', 'A_m4', 'B_m4', 'C_m4',
               'A_m5', 'B_m5', 'C_m5', 'speed']

target_corr = sample_df3[signal_cols + ['condition_encoded']].corr()['condition_encoded'].drop('condition_encoded').sort_values(ascending=False)

print("Correlation with Motor Condition (label encoded):")
print("Note: healthy=0, brb=1, swf=2 — correlation reflects label order, not fault severity")
display(target_corr)


# ==========================================
# FULL CORRELATION HEATMAP
# MCSA Dataset — Signal Columns Only
# ==========================================

# Use a sample to avoid memory issues with 2.7M rows
sample_df3 = df3.sample(n=50000, random_state=42)

signal_cols = ['A_m1', 'B_m1', 'C_m1', 'A_m2', 'B_m2', 'C_m2',
               'A_m3', 'B_m3', 'C_m3', 'A_m4', 'B_m4', 'C_m4',
               'A_m5', 'B_m5', 'C_m5']

plt.figure(figsize=(14, 10))
sns.heatmap(
    sample_df3[signal_cols].corr(),
    annot=True,
    fmt='.2f',
    cmap='coolwarm',
    center=0,
    linewidths=0.5,
    annot_kws={'size': 8}
)
plt.title('Signal Correlation Heatmap — MCSA Dataset\n(based on 50,000 sample rows)', 
          fontsize=13)
plt.tight_layout()
plt.show()
```
